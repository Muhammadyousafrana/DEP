# Import all the necessary library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

# Download the data by clicking the url given below:
* [UCI Online Retail || Dataset](https://archive.ics.uci.edu/dataset/352/online+retail)

```python
# load data and convert it into csv format
retail = pd.read_excel("./Online Retail.xlsx")
retail.to_csv("online_retail.csv", index=False)
```

```python
#load the coverted data and parse the date column
df = pd.read_csv("./online_retail.csv", low_memory=False, parse_dates=["InvoiceDate"])
df.head()
```

* separate the `year`
* separate the `Month`
* separate the `day` 
* separate the `day of week`
* separate the `day of year`
* drop the `InvoiceDate` column
#### we do separate all the stuff so that our model perfom well and find the pattern in the data

```python
df["InvoiceYear"] = df.InvoiceDate.dt.year
df["InvoiceMonth"] = df.InvoiceDate.dt.month
df["InvoiceDay"]  = df.InvoiceDate.dt.day
df["InvoiceDayOfWeek"] = df.InvoiceDate.dt.dayofweek
df["InvoiceDayOfYear"] = df.InvoiceDate.dt.dayofyear
df.drop("InvoiceDate", axis=1, inplace=True)
```

```python
#find total purchase and create a column field for it
df["TotalPurchase"] = df["Quantity"] * df["UnitPrice"]
df
```

```python
# Check for any missing value in the data
df.isna().sum()
```

* drop the missing `CustomerID` values

```python
df.dropna(subset=['CustomerID'], inplace=True)
```

```python
df["CustomerID"].isna().sum()
```

```python
# Aggregating data at customer level
customer_data = df.groupby('CustomerID').agg({
    'TotalPurchase': 'sum',
    'InvoiceNo': 'count',
    'Quantity': 'sum',
    'InvoiceYear': 'first',  # Assuming first year of transaction and same for other
    'InvoiceMonth': 'first',
    'InvoiceDay': 'first',  
    'InvoiceDayOfWeek': 'first',
    'InvoiceDayOfYear': 'first' 
}).reset_index()

# Renaming for clarity
customer_data.rename(columns={'InvoiceNo': 'PurchaseFrequency'}, inplace=True)

# Display the aggregated customer data
customer_data
```

```python
# Normalization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[['TotalPurchase', 'PurchaseFrequency', 'Quantity', 'InvoiceYear', 'InvoiceMonth', 'InvoiceDay', 'InvoiceDayOfWeek', 'InvoiceDayOfYear']])
```

```python
# Determine the optimal number of clusters using the Elbow Method
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()
plt.savefig("myplot.png")
```

## Optimat cluster should be 3 or 4 in the following case we're using `4` `Optimal cluster`

```python
# Clustering
optimal_clusters = 4
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
# Visualization
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
customer_data['PC1'] = principal_components[:, 0]
customer_data['PC2'] = principal_components[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=customer_data, palette='viridis')
plt.title('Customer Segments')
plt.show()
```

# Save the Segmented data to csv format

```python
customer_data.to_csv('customer_segments.csv', index=False)
```
