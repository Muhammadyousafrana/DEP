**LOADING DATA**

This loads the dataset into a DataFrame and prints the first few rows and the column names to understand the structure of the data.

```python
df = pd.read_csv("/content/tiktok_google_play_reviews.csv")
print(df.columns)
df.head()
```

**Handling Missing Values**  

This part checks for missing values and drops rows where the 'content' or 'score' columns are missing, ensuring data integrity.

**Data Preprocessing**

Text data is cleaned by converting to lowercase, removing punctuation, and eliminating stopwords to prepare it for analysis.

```python
# Check for missing values
df.isnull().sum()

# Drop missing values
df.dropna(subset=['content', 'score'], inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

# Convert text to lowercase
df['content'] = df['content'].str.lower()

# Remove punctuation
df['content'] = df['content'].str.replace('[^\w\s]', '')

# Remove stopwords
stop = stopwords.words('english')
df['content'] = df['content'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

# Display the cleaned data
df.head()
```

Granular Sentiment Categorization

Sentiments are further categorized into more granular classes, providing a detailed view of sentiment distribution.

```python
# Define a function to categorize sentiment into more granular classes
def categorize_granular_sentiment(score):
    if score == 5:
        return 'very positive'
    elif score == 4:
        return 'positive'
    elif score == 3:
        return 'neutral'
    elif score == 2:
        return 'negative'
    else:
        return 'very negative'

# Apply the function to the score column
df['granular_sentiment'] = df['score'].apply(categorize_granular_sentiment)

# Display the distribution of granular sentiments
sns.countplot(df['granular_sentiment'])
plt.title('Distribution of Granular Sentiment Labels')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.show()
```
